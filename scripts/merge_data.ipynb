```python
methylation_df = pd.read_csv("../data/data_methylation_promoters_rrbs.txt", sep="\t")
# methylation_df.T
methylation_df = methylation_df.set_index('Hugo_Symbol').T.reset_index().rename(columns={'index': 'PATIENT_ID'}).copy()
# methylation_df.columns.name = None
print(methylation_df.columns)
print(len(set(methylation_df.columns)))
print(len(methylation_df.columns))
methylation_df
```

```python
# show duplicate columns in methylation_df

methylation_df[methylation_df.columns[methylation_df.columns.duplicated()].unique().tolist()]
```

```python
# Handle duplicate columns by adding suffixes
duplicate_cols = methylation_df.columns[methylation_df.columns.duplicated()].unique().tolist()

# Create a dictionary to track column occurrences
col_counts = {}
new_columns = []
renamed_cols = []

for col in methylation_df.columns:
    if col in col_counts:
        col_counts[col] += 1
        new_columns.append(f"{col}_{col_counts[col]}")
        renamed_cols.append(f"{col}_{col_counts[col]}")
    else:
        col_counts[col] = 0
        new_columns.append(col)

methylation_df.columns = new_columns
print(f"Duplicate columns renamed: {duplicate_cols}\n{renamed_cols}")
methylation_df
```

```python
# Handle duplicate columns by adding suffixes
duplicate_cols = microarray_df.columns[microarray_df.columns.duplicated()].unique().tolist()

# Create a dictionary to track column occurrences
col_counts = {}
new_columns = []
renamed_cols = []

for col in microarray_df.columns:
    if col in col_counts:
        col_counts[col] += 1
        new_columns.append(f"{col}_{col_counts[col]}")
        renamed_cols.append(f"{col}_{col_counts[col]}")
    else:
        col_counts[col] = 0
        new_columns.append(col)

microarray_df.columns = new_columns
print(f"Duplicate columns renamed: {duplicate_cols}\n{renamed_cols}")
microarray_df
```

```python
from matplotlib import pyplot as plt
from upsetplot import UpSet, from_memberships

# Get patient IDs from each dataset
patient_ids = set(patient_df['PATIENT_ID'])
methylation_ids = set(methylation_df['PATIENT_ID'])
microarray_ids = set(microarray_df['PATIENT_ID'])

# Create membership list for UpSet plot
memberships = []

# Patients in all three datasets
all_three = patient_ids & methylation_ids & microarray_ids
memberships.extend([('Patient', 'Methylation', 'Microarray')] * len(all_three))

# Patients in patient and methylation only
patient_meth = (patient_ids & methylation_ids) - microarray_ids
memberships.extend([('Patient', 'Methylation')] * len(patient_meth))

# Patients in patient and microarray only
patient_micro = (patient_ids & microarray_ids) - methylation_ids
memberships.extend([('Patient', 'Microarray')] * len(patient_micro))

# Patients in methylation and microarray only
meth_micro = (methylation_ids & microarray_ids) - patient_ids
memberships.extend([('Methylation', 'Microarray')] * len(meth_micro))

# Patients in patient only
patient_only = patient_ids - methylation_ids - microarray_ids
memberships.extend([('Patient',)] * len(patient_only))

# Patients in methylation only
meth_only = methylation_ids - patient_ids - microarray_ids
memberships.extend([('Methylation',)] * len(meth_only))

# Patients in microarray only
micro_only = microarray_ids - patient_ids - methylation_ids
memberships.extend([('Microarray',)] * len(micro_only))

# Create UpSet plot
upset_data = from_memberships(memberships)
upset = UpSet(upset_data, subset_size='count', show_counts=True)
upset.plot()
plt.suptitle('Intersection of Patient IDs Across Datasets', y=1.02)
plt.tight_layout()
plt.show()

# Print summary statistics
print(f"Total unique patients in patient_df: {len(patient_ids)}")
print(f"Total unique patients in methylation_df: {len(methylation_ids)}")
print(f"Total unique patients in microarray_df: {len(microarray_ids)}")
print(f"Patients in all three datasets: {len(all_three)}")
```

```python
from matplotlib_venn import venn3

# Create Venn diagram
plt.figure(figsize=(10, 8))
venn3(
    [patient_ids, methylation_ids, microarray_ids],
    set_labels=('Patient', 'Methylation', 'Microarray')
)
plt.title('Intersection of Patient IDs Across Datasets')
plt.tight_layout()
plt.show()
```
